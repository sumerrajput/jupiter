```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
```

```python
x=('https://www.flipkart.com/search?q=apple&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
```

```python
rp=requests.get(x)
```

```python
rp.status_code
```

```python
soup=BeautifulSoup(rp.content,'lxml')
```

```python
print(soup.prettify())
```

```python
Name=[]
Rating=[]
Orignal_Price=[]
Discount=[]
After_Discount=[]


#loop

for i in results:
    #name
    try:
        Name.append(i.find('div',{'class':"_4rR01T"}).text)
    except:
        Name.append('n/a')
    
    try:
        Rating.append(i.find('span',{'class':'_2_R_DZ'}).text)
    except:
        Rating.append('n/a')
    
    try:
        Orignal_Price.append(i.find('div',{'class':'_3I9_wc _27UcVY'}).text)
    except:
        Orignal_Price.append('n/a')
        
    try:
        Discount.append(i.find('div',{'class':'_3Ay6Sb'}).text)
    except:
        Discount.append('n/a')
        
    try:
        After_Discount.append(i.find('div',{'class':'_30jeq3 _1_WHN1'}).text)
    except:
        After_Discount.append('n/a')
        
    
        
        
        
    
 
```

# create df

```python
AD=pd.DataFrame({'NAME':Name,'Rating':Rating,'Orignal Price':Orignal_Price,
          "Discount":Discount,'After Discount':After_Discount    })
```

```python
AD['Rating']=AD['Rating'].apply(lambda x:x.split(" ")[0])
```

```python
AD.to_excel('Apple_Data.xlsx',index=False)
```

# now for number of pages

```python
Name=[]
Rating=[]
Orignal_Price=[]
Discount=[]
After_Discount=[]


for j in range(1,11):
    x=('https://www.flipkart.com/search?q=apple&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off%27%29&page={}'.format(j))
    
    rp=requests.get(x)
    
    soup=BeautifulSoup(rp.content,'lxml')

    results=soup.find_all('a',{'class':'_1fQZEK'})
    
    for i in results:
    #name
        try:
            Name.append(i.find('div',{'class':"_4rR01T"}).text)
        except:
            Name.append('n/a')

        try:
            Rating.append(i.find('span',{'class':'_2_R_DZ'}).text)
        except:
            Rating.append('n/a')

        try:
            Orignal_Price.append(i.find('div',{'class':'_3I9_wc _27UcVY'}).text)
        except:
            Orignal_Price.append('n/a')

        try:
            Discount.append(i.find('div',{'class':'_3Ay6Sb'}).text)
        except:
            Discount.append('n/a')

        try:
            After_Discount.append(i.find('div',{'class':'_30jeq3 _1_WHN1'}).text)
        except:
            After_Discount.append('n/a')
    
```

```python
AD=pd.DataFrame({'NAME':Name,'Rating':Rating,'Orignal Price':Orignal_Price,
          "Discount":Discount,'After Discount':After_Discount    })
```
